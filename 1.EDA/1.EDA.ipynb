## 1.1 Measuring the central tendency

### 1. mean

#### Arithmetic mean

```python
# Importing libraries
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# Load dataset
index = [1,2,3,4,5]
age = [31, 22, 43, 24, 35] 
```

```python
# Graph details
plt.xlabel("Index")
plt.ylabel("Age")
plt.title("Employees Age")
# Plot graph
plt.plot(index, age, color="#880066", marker="D", linewidth=3, markersize=10)     
plt.tight_layout()
plt.show()
```

```python
# Arithmetic mean of age variable
import numpy as np
age = np.array(age)
print("Arithmetic mean : " ,age.mean())
```

#### Weighted arithmetic mean

```python
# Weighted arithmetic mean of age variable
age = [31, 22, 43, 24, 35] 
weight = [0.2, 0.9, 0.4, 0.3, 0.6] 

numerator = 0
for i in [0,1,2,3,4]:
    numerator += age[i]*weight[i] 
denominator = sum(weight)

print("Weighted arithmetic mean : " ,numerator/denominator)
```

#### Geometric mean

```python
# Load dataset
month = [1,2,3,4,5]
profit = [3,9,27,81,243] 
```

```python
# Graph details
plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("Company profit")
# Plot graph
plt.plot(month, profit, color="#880066", marker="D", linewidth=3, markersize=10)     
plt.tight_layout()
plt.show() 
```

```python
# Arithmetic mean of age variable
import numpy as np
profit = np.array(profit)
print("Arithmetic mean : " ,profit.mean())
```

```python
# Geometric mean
from scipy import stats
print("Geometric mean : " ,stats.gmean(profit)) 
```

#### Harmonic mean

```python
# Load dataset
month = [1,2,3,4,5]
profit = [13, 9, 12, 81, 10] 
```

```python
# Graph details
plt.xlabel("Month")
plt.ylabel("Profit")
plt.title("Company profit")
# Plot graph
plt.plot(month, profit, color="#880066", marker="D", linewidth=3, markersize=10)     
plt.tight_layout()
plt.show() 
```

```python
# Arithmetic mean of age variable
import numpy as np
profit = np.array(profit)
print("Arithmetic mean : " ,profit.mean())
```

```python
# Geometric mean
from statistics import geometric_mean
print("Geometric mean : " ,geometric_mean(profit)) 
```

```python
# Harmonic mean
import statistics
print("Harmonic Mean : ", statistics.harmonic_mean(profit))
```

## 2.2 Covariance plot

```python
# Load random set of values
data = pd.read_csv("E:\\MY LECTURES\\DATA SCIENCE\\3.Programs\\dataset\\Meal and Tips.csv") 
data.head()
```

```python
plt.boxplot(data["size"])
```

```python
# Covariance calculation
x = np.array(data["total_bill"])
y = np.array(data["tip"])
covariance = np.cov(x, y)[0][1]
print(covariance)
```

```python
# Draw scatter plot  
plt.scatter(x,y) 
plt.show() 
```

## 2.3 Correlation

```python
# Correlation calculation
x = np.array(data["total_bill"])
y = np.array(data["tip"])
covariance = np.corrcoef(x, y)[0][1]
print(covariance)
```

```python
# Draw scatter plot  
plt.scatter(x,y) 
plt.show()  
```
